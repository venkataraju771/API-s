```python
import requests

API_URL = "https://samples.openweathermap.org/data/2.5/forecast/hourly?q=London,us&appid=b6907d289e10d714a6e88b30761fae22"


def get_weather_data():
    response = requests.get(API_URL)
    data = response.json()
    return data["list"]


def get_temperature(date):
    weather_data = get_weather_data()
    for entry in weather_data:
        if entry["dt_txt"].startswith(date):
            return entry["main"]["temp"] - 273.15  # Convert from Kelvin to Celsius

    return None


def get_wind_speed(date):
    weather_data = get_weather_data()
    for entry in weather_data:
        if entry["dt_txt"].startswith(date):
            return entry["wind"]["speed"]

    return None


def get_pressure(date):
    weather_data = get_weather_data()
    for entry in weather_data:
        if entry["dt_txt"].startswith(date):
            return entry["main"]["pressure"]

    return None


def main():
    while True:
        print("1. Get weather")
        print("2. Get Wind Speed")
        print("3. Get Pressure")
        print("0. Exit")

        option = int(input("Enter your choice: "))

        if option == 1:
            date = input("Enter the date (YYYY-MM-DD): ")
            temperature = get_temperature(date)
            if temperature is not None:
                print(f"Temperature on {date}: {temperature:.2f}°C")
            else:
                print("Data not available for the input date.")
        elif option == 2:
            date = input("Enter the date (YYYY-MM-DD): ")
            wind_speed = get_wind_speed(date)
            if wind_speed is not None:
                print(f"Wind Speed on {date}: {wind_speed} m/s")
            else:
                print("Data not available for the input date.")
        elif option == 3:
            date = input("Enter the date (YYYY-MM-DD): ")
            pressure = get_pressure(date)
            if pressure is not None:
                print(f"Pressure on {date}: {pressure} hPa")
            else:
                print("Data not available for the input date.")
        elif option == 0:
            print("Exiting the program.")
            break
        else:
            print("Invalid option. Please try again.")


if __name__ == "__main__":
    main()
```
